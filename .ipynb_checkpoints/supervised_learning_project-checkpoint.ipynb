# Part II : Preprocessing & Feature Engineering

You need to perform preprocessing on the given dataset. Please consider the following tasks and carry out the necessary steps accordingly.
- Handling missing values
- Handling outliers
- Scaling and normalization
- Feature Engineering
- Handling imbalanced data

```python
data = pd.read_csv('diabetes.csv')
# Pregnancies: remove high outliers
data = data[data['Pregnancies'] <= (data.describe()['Pregnancies']['75%'] + 1.5*(data.describe()['Pregnancies']['75%'] - data.describe()['Pregnancies']['25%']))]
# Glucose: possible low outliers / 0 values
data = data[data['Glucose']!=0]
data = data[data['Glucose'] <= (data.describe()['Glucose']['75%'] + 1.5*(data.describe()['Glucose']['75%'] - data.describe()['Glucose']['25%']))]
# Blood Pressure: possible high outliers, 0 values
data = data[data['BloodPressure']!=0]
data = data[data['BloodPressure'] <= (data.describe()['BloodPressure']['75%'] + 1.5*(data.describe()['BloodPressure']['75%'] - data.describe()['BloodPressure']['25%']))]
data = data[data['BloodPressure'] >= (data.describe()['BloodPressure']['25%'] - 1.5*(data.describe()['BloodPressure']['75%'] - data.describe()['BloodPressure']['25%']))]
# Skin Thickness: 0 values
# Many 0 values - consider excluding column
data = data[data['SkinThickness']!=0]
data = data[data['SkinThickness'] <= (data.describe()['SkinThickness']['75%'] + 1.5*(data.describe()['SkinThickness']['75%'] - data.describe()['SkinThickness']['25%']))]
data = data[data['SkinThickness'] >= (data.describe()['SkinThickness']['25%'] - 1.5*(data.describe()['SkinThickness']['75%'] - data.describe()['SkinThickness']['25%']))]
# data['SkinThickness'].value_counts()
# data = data.drop('SkinThickness', axis=1)
# Insulin: low & high outliers
data = data[data['Insulin']!=0]
data = data[data['Insulin'] <= (data.describe()['Insulin']['75%'] + 1.5*(data.describe()['Insulin']['75%'] - data.describe()['Insulin']['25%']))]
data = data[data['Insulin'] >= (data.describe()['Insulin']['25%'] - 1.5*(data.describe()['Insulin']['75%'] - data.describe()['Insulin']['25%']))]
# BMI: low & high outliers
data = data[data['BMI']!=0]
data = data[data['BMI'] <= (data.describe()['BMI']['75%'] + 1.5*(data.describe()['BMI']['75%'] - data.describe()['BMI']['25%']))]
data = data[data['BMI'] >= (data.describe()['BMI']['25%'] - 1.5*(data.describe()['BMI']['75%'] - data.describe()['BMI']['25%']))]
# DiabetesPedigreeFunction: possible high outliers
data = data[data['DiabetesPedigreeFunction']!=0]
data = data[data['DiabetesPedigreeFunction'] <= (data.describe()['DiabetesPedigreeFunction']['75%'] + 1.5*(data.describe()['DiabetesPedigreeFunction']['75%'] - data.describe()['DiabetesPedigreeFunction']['25%']))]
data = data[data['DiabetesPedigreeFunction'] >= (data.describe()['DiabetesPedigreeFunction']['25%'] - 1.5*(data.describe()['DiabetesPedigreeFunction']['75%'] - data.describe()['DiabetesPedigreeFunction']['25%']))]
# Age: normal but high outliers
data = data[data['Age']!=0]
data = data[data['Age'] <= (data.describe()['Age']['75%'] + 1.5*(data.describe()['Age']['75%'] - data.describe()['Age']['25%']))]
data = data[data['Age'] >= (data.describe()['Age']['25%'] - 1.5*(data.describe()['Age']['75%'] - data.describe()['Age']['25%']))]
# Outcome: assume normal, reevalute following cleaning
```

# Part III : Training ML Model

For this task, you are required to build a machine learning model to predict the outcome variable. This will be a binary classification task, as the target variable is binary. You should select at least two models, one of which should be an ensemble model, and compare their performance.

- Train the models: Train the selected models on the training set.
- Model evaluation: Evaluate the trained models on the testing set using appropriate evaluation metrics, such as accuracy, precision, recall, F1-score, and ROC-AUC.
- Model comparison: Compare the performance of the selected models and choose the best-performing model based on the evaluation metrics. You can also perform additional analysis, such as model tuning and cross-validation, to improve the model's performance.

```python
# Pipeline to perform scaling, principal component analysis, and logistic regression.
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

X = data.drop(columns='Outcome')
y = data['Outcome']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=16, stratify=y) #stratify is nice to use in case of class imbalance
```

```python
from sklearn.pipeline import Pipeline

pipeline = Pipeline(steps=[('scaling', StandardScaler()),
                           ('pca', PCA(n_components=3)),
                           ('classifier', LogisticRegression())])
pipeline.fit(X_train, y_train)

y_pred = pipeline.predict(X_test)
acc = accuracy_score(y_test, y_pred)
print(f'Test set accuracy: {acc}')
```

```python
from sklearn import set_config
set_config(display='diagram')
pipeline
```

```python
from sklearn.pipeline import FeatureUnion
from sklearn.feature_selection import SelectKBest
from sklearn.linear_model import RidgeClassifier
from sklearn.model_selection import GridSearchCV

feature_union = FeatureUnion([('pca', PCA()),
                              ('select_best', SelectKBest())])

pipeline = Pipeline(steps=[('scaling', StandardScaler()),
                           ('features', feature_union),
                           ('classifier', RidgeClassifier())])

# Find the best hyperparameters using GridSearchCV on the train set
param_grid = {'classifier__alpha': [0.001, 0.005, 0.01, 0.05, 0.1, 0.5],
              'features__pca__n_components': [2, 3, 4, 5],
              'features__select_best__k': [2, 3, 4, 5]}
grid = GridSearchCV(pipeline, param_grid=param_grid, cv=5)
grid.fit(X_train, y_train)

best_model = grid.best_estimator_
best_hyperparams = grid.best_params_
best_acc = grid.score(X_test, y_test)
print(f'Best test set accuracy: {best_acc}\nAchieved with hyperparameters: {best_hyperparams}')
```

```python
from sklearn import set_config
set_config(display='diagram')
pipeline
```

```python
from sklearn.svm import SVC

pipeline = Pipeline(steps=[('scaling', StandardScaler()),
                           ('features', feature_union),
                           ('classifier', LogisticRegression())])

# Find the best hyperparameters and model using GridSearchCV on the train set
param_grid = {'classifier': [RidgeClassifier(alpha=0.1), RidgeClassifier(alpha=0.01), SVC()],    # Which is better, Logistic Regression or an SVM Classifier?
              'features__pca__n_components': [2, 3, 4, 5],
              'features__select_best__k': [2, 3, 4, 5]}
grid = GridSearchCV(pipeline, param_grid=param_grid, cv=5)
grid.fit(X_train, y_train)

best_model = grid.best_estimator_
best_hyperparams = grid.best_params_
best_acc = grid.score(X_test, y_test)
print(f'Best test set accuracy: {best_acc}\nAchieved with hyperparameters: {best_hyperparams}')
```

```python
from sklearn import set_config
set_config(display='diagram')
pipeline
```

```python
from sklearn.pipeline import Pipeline

pipeline = Pipeline(steps=[('scaling', StandardScaler()),
                           ('pca', PCA(n_components=5)),
                           ('classifier', RidgeClassifier(alpha=0.1))])
pipeline.fit(X_train, y_train)

y_pred = pipeline.predict(X_test)
acc = accuracy_score(y_test, y_pred)
print(f'Test set accuracy: {acc}')
```
